```python
import numpy as np
import matplotlib.pyplot as plt

r_data = np.loadtxt("Sullivan_PhysGP_Ar_r.txt")  # columns: r, mu_r, 2*std_r

plt.figure(figsize=(6, 4), dpi=300)
plt.title("Argon RDF Posterior")
plt.plot(r_data[:, 0], r_data[:, 1], label=r'$\mu_r$', color='C0')
plt.fill_between(r_data[:, 0], r_data[:, 1] - r_data[:, 2], r_data[:, 1] + r_data[:, 2],
                 alpha=0.3, label=r'$\pm2\sigma_r$', color='C0')
plt.xlabel(r"$r$")
plt.ylabel(r"$g(r)$")
plt.xlim(0, 12)
plt.legend()
plt.tight_layout()
plt.show()

q_data = np.loadtxt("Sullivan_PhysGP_Ar_q.txt")  # columns: q, mu_q, 2*std_q

plt.figure(figsize=(6, 4), dpi=300)
plt.title("Argon S(q) Posterior")
plt.plot(q_data[:, 0], q_data[:, 1], label=r'$\mu_q$', color='C1')
plt.fill_between(q_data[:, 0], q_data[:, 1] - q_data[:, 2], q_data[:, 1] + q_data[:, 2],
                 alpha=0.3, label=r'$\pm2\sigma_q$', color='C1')
plt.xlabel(r"$q$")
plt.ylabel(r"$S(q)$")
plt.xlim(0, 12)
plt.legend()
plt.tight_layout()
plt.show()

train_data = np.loadtxt("Sullivan_PhysGP_Ar_q_training.txt")  # columns: q, perturbed_sq, yarnell_sq

plt.figure(figsize=(6, 4), dpi=300)
plt.title("Training Data for S(q)")
plt.plot(train_data[:, 0], train_data[:, 1], 'o', markersize=1, alpha=0.6,
         color='black', label="Perturbed S(q)")
plt.plot(train_data[:, 0], train_data[:, 2], '--', markersize=1, alpha=0.6,
         color='black', label="Yarnell S(q)")
plt.xlabel(r"$q$")
plt.ylabel(r"$S(q)$")
plt.xlim(0, 12)
plt.legend()
plt.tight_layout()
plt.show()

cov_r = np.loadtxt("Sullivan_PhysGP_Ar_cov_r.txt")
r_vals = r_data[:, 0]

plt.figure(figsize=(5, 4), dpi=300)
plt.title("Argon Covariance Matrix _r")
plt.imshow(cov_r, cmap="viridis", origin="lower", extent=[r_vals[0], r_vals[-1], r_vals[0], r_vals[-1]])
plt.xlabel(r"$r$ $[\AA]$")
plt.ylabel(r"$r'$ $[\AA]$")
plt.colorbar(label="Covariance")
plt.tight_layout()
plt.show()

cov_q = np.loadtxt("Sullivan_PhysGP_Ar_cov_q.txt")
q_vals = q_data[:, 0]

plt.figure(figsize=(5, 4), dpi=300)
plt.title("Argon Covariance Matrix _q")
plt.imshow(cov_q, cmap="viridis", origin="lower", extent=[q_vals[0], q_vals[-1], q_vals[0], q_vals[-1]])
plt.xlabel(r"$q$ $[\AA^{-1}]$")
plt.ylabel(r"$q'$ $[\AA^{-1}]$")
plt.colorbar(label="Covariance")
plt.tight_layout()
plt.show()
```
